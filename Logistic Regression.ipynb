```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np
from sklearn.model_selection import train_test_split
import seaborn as sns
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
url = "https://rb.gy/9wttb"
names = ['preg','plas','skin','test','mass','pedi','age','class']
data = pd.read_csv(url,names=names)
```

```python
# To see the table:
data
```

```python
# Print them and see the details:
print(data.head())
print(data.describe())
print(data.info())
```

```python
sns.pairplot(data,hue='class')
plt.show()
```

```python
# Split the Features(x) and the Target (y):
# By using iloc :
# Features :
# Make it parmently:
x=data.iloc[:,:-1]
x
```

```python
# Target :
# Make it parmently:
y=data.iloc[:,-1]
y
```

```python
# OR :
X = data.drop('class',axis=1)
Y = data['class']
```

```python
# Tarin and Test the data :
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
# Create a Logistic Regression object and train it using the training data :
lr = LogisticRegression(max_iter=400)
lr.fit(x_train,y_train)
```

```python
# Make predictions using test set:
y_pred = lr.predict(x_test)
```

```python
# Evaluate the model performance of the algorithm :
print('Accuarcy:' , metrics.accuracy_score(y_test,y_pred) )
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
