# Temperature vs. Energy Consumption :

```python
# The dataset consists of daily average temperatures (in Celsius) and corresponding energy consumption
# (in kilowatt-hours) in a residential building. Each data point represents a single day.
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

# Objective

```python
# the data as csv and read the DataFrame:
df = pd.read_csv('Ass2.csv')
df
```

```python
# Splite the dataset into input features (x) and Target variable (y)

x=df[['Temperature']]
y=df[['Energy Compumption']]
```

```python
# Build and train linear regression model
model = LinearRegression()
```

```python
# Fit the model to the data 
model.fit(x,y)
```

```python
# Predict Energy Compumption based on (17,26,32)
Energy = [[17],[26],[32]]
predicted_scores = model.predict(Energy)

for EnergyC, scores in zip(Energy, predicted_scores):
    print(f"daily average Energy Compumption: {EnergyC[0]}, predicted Test scores:{scores[0]:.2f}" )
```

```python
# Set the theme:
sns.set_theme(style='whitegrid')

# Create a Scatter plot with data points
sns.scatterplot(data=df, x='Temperature', y='Energy Compumption',label='Data Points')

#Create a regression line
sns.regplot(data=df, x='Temperature', y='Energy Compumption',scatter=False, color='steelblue',label='Regression Line')

#Predict the Energy Compumption for Energy
Energy = [[17],[26],[32]]
predicted_scores = model.predict(Energy)

# Plot the predicted values
sns.scatterplot(x=[x[0] for x in Energy], y=[y[0] for y in predicted_scores], color='red', marker='s', s=100, label='Predicted Values')

# Set plot title and labels
plt.title('Temperature vs. Energy Compumption')
plt.xlabel('Temperature')
plt.ylabel('Energy Compumption')

# Add a legand
plt.legend()

# Display the plot
plt.show()
```
