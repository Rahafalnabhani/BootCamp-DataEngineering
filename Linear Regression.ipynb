```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create a DataFrame from DataSet:
data = {'Hours Studied':[2,3,4,5,6],
        'Test Score':[60,70,80,90,95]}
df = pd.DataFrame(data)
```

```python
# Splite the dataset into input features (x) and Target variable (y)
# This steps mean: Training between then Testing ---> Split

x = df[['Hours Studied']]
y = df[['Test Score']]
```

```python
# Create Linear Regression object
# Build the model
model = LinearRegression()
```

```python
# Fit the model to the data
# Take the training Data which is x

model.fit(x,y)
```

```python
# Predict the test scores for new student 
# Test the data 
# Two x 3.5 , 4.5
new_hours = [[3.5],[4.5]]
predicted_scores = model.predict(new_hours)

# Print the predicted scores
# Loop
# The function mean that will take them at the same time (new_hours, predicted_scores)
# f mean there is format 
# {two variable}
# Only using List and Array
# Socres is float and have round the numbers [.2f] decimal numbers 2digits 

for hours, scores in zip(new_hours, predicted_scores):
  print(f"Hours Studied: {hours[0]}, predicted Test Scores: {scores[0]:.2f}")
```

```python
# Set a Beautiful Seaborn theme
# Set the theme of the darwing
# The real data 
sns.set_theme(style='whitegrid')

# Create a Scatter plot with data points
sns.scatterplot(data=df, x='Hours Studied', y='Test Score',label='Data Points')

#Create a regression line
sns.regplot(data=df, x='Hours Studied', y='Test Score',scatter=False, color='steelblue',label='Regression Line')

#Predict the test scores for new houres studied
new_hours = [[3.5],[4.5]]
predicted_scores = model.predict(new_hours)

# Plot the predicted values
sns.scatterplot(x=[x[0] for x in new_hours], y=[y[0] for y in predicted_scores], color='red', marker='s', s=100, label='Predicted Values')

# Set plot title and labels
plt.title('Hours Studied vs. Test Score')
plt.xlabel('Hours Studied')
plt.ylabel('Test Scores')

# Add a legand
plt.legend()

# Display the plot
plt.show()
```
